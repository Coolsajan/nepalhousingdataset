```python
#importing libaries
import pandas as pd
```

```python
#Loading first dataset
df_basbobas=pd.read_csv('baobas/preprocessed_data_basobas.csv')
```

```python
df_basbobas.info()
```

```python
df_basbobas=df_basbobas.rename(columns={'floors':'no_of_flat'})
df_basbobas.head()
```

```python
df_ghadgaderi=pd.read_csv('gharghaderi/preprocessed_ghargaderi.csv')
df_ghadgaderi=df_ghadgaderi.rename(columns={"no_of_parking":"parking","land_area":"area"})
df_ghadgaderi.head()
```

```python
df_kantipur=pd.read_csv('kantipur_real_estate/preprocessed_kantipur.csv')
df_kantipur=df_kantipur.rename(columns={"land_area":"area","furnishing":"furniture"})
df_kantipur.head()
```

```python
working_dataset=pd.concat([df_basbobas,df_ghadgaderi,df_kantipur])
```

```python
working_dataset.head()
```

```python
working_dataset.to_csv('nepal_hosuing_datetset.csv',index=False)
```
